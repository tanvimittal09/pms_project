```python
#Importing required header files...

#importing tkinter and it's required submodules...
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import tkinter as tk

#importing datetime module for time and date on ui and for getting system time for vehicle entry...
import datetime

#Importing DateEntry module for letting the admin choose custom date interactively...
#please install this module using
#pip install tkcalendar
from tkcalendar import DateEntry

import mysql.connector

import math


#################################################################################################
###---------------------Database Design-------------------------------------------------------###
###---------------------------------------By:- Deepti Mehra-----------------------------------###
#################################################################################################

#Connecting to the database
#Please change Password as per your database...
mydb= mysql.connector.connect(host='localhost', user='root', password='root')

#Create a cursor myC of mydb...
myC = mydb.cursor(buffered=True)

#Statements to be fired in the starting
            #Creates a database PMS if it does not exist on the system...
sql_stuff=("CREATE DATABASE IF NOT EXISTS PMS;",
           #Start using this database...
           "USE PMS;",
           #Create table vehcles_parked if it does not exists to store the information of the vehicles currently parked...
           "CREATE TABLE IF NOT EXISTS Vehicles_Parked (no varchar(15), type VARCHAR(20), model VARCHAR(30), color VARCHAR(30), driver VARCHAR(30), mobile VARCHAR(10), entry TIMESTAMP, PRIMARY KEY(no));",
           #Create a table vehicle_log if it does not exist to store the information of all vehicles parked till date... 
           "CREATE TABLE IF NOT EXISTS Vehicle_log (log_id INTEGER AUTO_INCREMENT, no varchar(15), type VARCHAR(20), model VARCHAR(30), color VARCHAR(30), driver VARCHAR(30), mobile VARCHAR(10), entry TIMESTAMP, ext TIMESTAMP, PRIMARY KEY(log_id));"
          )

#Executing the above stated commands...
for command in sql_stuff:
    myC.execute(command)
    
#And committing the operations done...
mydb.commit()

#A dictionary for storing the parking charges per hour associated with each vehicle type...
price_per_hour={
    'Cycle/ Rickshaw' : 10,
    'Bike/ Scooter'   : 20,
    'Auto-Rickshaw'   : 30,
    'Car'             : 40,
    'Bus/ Truck'      : 50          
}
    
#################################################################################################
###---------------------Homescreen Module-----------------------------------------------------###
###---------------------------------------By:- Tanvi Mittal-----------------------------------###
#################################################################################################

#Home screen module which is invoked when user signs in using right credentials...
def homeScreen():
    
    #Function destroying the current window and calling vehicleEntry module and calling itself again when vehicleEntry is closed...
    def vEn():
        root.destroy()
        vehicleEntry()
        homeScreen()
        
    #Function destroying the current window and calling vehicleExit module and calling itself again when vehicleExit is closed...    
    def vEx():
        root.destroy()
        vehicleExit()
        homeScreen()
        
    #Main root window    
    root=Tk()
    
    #Setting root geometry to approximately full screen size...
    root.geometry("{0}x{1}+0+0".format(root.winfo_screenwidth(), root.winfo_screenheight()))
    
    #Configuring the background color...
    root.config(background="#0da2ff")
    
    #Creating canvas with h=400 and w=400...
    c=Canvas(root,height=400,width=400)
    #packing it by letting it expand it without space restriction...
    c.pack(expand="True")
    
    #Creating main heading label...
    l1=Label(c,text="PARKING MANAGEMENT SYSTEM",font=("aerial",18,"bold"),bg="red",fg="white")
    l1.place(x=200,y=50,anchor="center")
    
    #Creating sub heading...
    l2=Label(c,text="Choose an appropriate option :",font=("aerial",16))
    l2.place(x=60,y=90)
    
    #Button invoking vEn function...
    b3=Button(root,text="1)Vehicle Entry",font="6", command=vEn)
    b3.place(x=730,y=350,anchor="center")
    
    #Button invoking vEx function...
    b4=Button(root,text="2)Vehicle Exit",font="6", command=vEx)
    b4.place(x=720,y=400,anchor="center")
    
    #Button calling parkedVehicles module...
    b5=Button(root,text="3)View Parked Vehicles",font="6", command=parkedVehicles)
    b5.place(x=760,y=450,anchor="center")
    
    #Button calling vehicle log module...
    b6=Button(root,text="4)View Parking Log",font="6", command=vehicleLog)
    b6.place(x=740,y=500,anchor="center")
    
    #Button to close this module...
    b7=Button(root,text="5)Exit",font="6", command=root.destroy)
    b7.place(x=680,y=550,anchor="center")
    
    #Starting mainloop...
    root.mainloop()
    
#################################################################################################
###---------------------Vehicle log Module----------------------------------------------------###
###---------------------------------------By:- Tanvi Mittal-----------------------------------###
#################################################################################################

#Vehicle log module...
def vehicleLog():
    #root window...
    root=Tk()
    #Heading
    L1=Label(root,text="VEHICLE LOG",font=("aerial",18,"bold"))
    L1.pack()
    #Creating listbox...
    lb=Listbox(root,fg="yellow",bg="grey", height=100, width=500,font="Helvetica")
    #Running sql command to get all records...
    sql="SELECT * FROM vehicle_log"
    myC.execute(sql)
    
    #Putting the fetched material in a variable result...
    result=myC.fetchall()
    
    #Iteratively, insert the records into the listBox...
    for x in result:
        lb.insert(0,x)
    
    #Pack the listbox
    lb.pack()
    
    #Mainloop
    root.mainloop()
    
#################################################################################################
###---------------------Sign-In Module--------------------------------------------------------###
###---------------------------------------By:- Milandeep Singh--------------------------------###
#################################################################################################

#Sign in module which takes username and password as input and gives access to the system...
def signIn():
    #This module verifies whether the credentials are correct or not...
    def verify(event):
        #If username and password are admin and password respectively,
        if e1.get()=='admin' and e2.get()=='password':
            #Then, close the current window
            root.destroy()
            #and call homeScreen Module...
            homeScreen()
        #Otherwise,
        else:
            #Show an error message that the credentials are incorrect...
            l5=Label(w,text='Wrong username or password!', bg='white', fg='red', font=('Helvetica', 10, 'bold'))
            l5.place(x=50, y=355)

    #Initialization of tkinter window for the module
    root=Tk()
    #Setting geometry of the module window...
    root.geometry('800x500')
    #Setting the background of root...
    root.config(bg='#0da2ff')
    #Setting title...
    root.title('Sign-In Window')
    
    #Creating a canvas inside root to put all other widgets inside
    w = Canvas(root, width=300,
                     height=400,
                     borderwidth=5,
                     background='white'
                     )
    #setting border thickness of the same
    w.config(highlightthickness='0')
    #packing tha canvas
    w.pack(padx=10,pady=40, expand=True)
    
    #Main heading of the module
    l1=Label(w,text="Sign In", fg='black', bg='white')
    l1['font']=('Helvetica', 20, 'bold')
    l1.place(x=110,y=30)
    
    #An exit button on the top right corner of the window with appropriate options
    exit_btn=Button(w,text='×',bg='white', relief=FLAT, width=1, height=1, command=root.destroy, font=('Helvetica', 12, 'bold'))
    exit_btn.place(x=290,y=2)
    exit_btn.config(highlightthickness='0', cursor='hand2')

    #Label asking for username...
    l2=Label(w,text="Username ", fg='black', bg='white', font=('Helvetica', 12))
    l2.place(x=15,y=90)

    #Textbox for username
    e1=Entry(w, font=('Helvetica', 10))
    e1.config(highlightbackground='black',highlightthickness='.5',width=45)
    e1.place(x=20,y=120,width=270, height=30)

    #Displaying default username
    l4=Label(w,text='Default username : admin', bg='white', font=('Arial',8))
    l4.place(x=15,y=150)

    #Label asking for password...
    l3=Label(w,text="Password", fg='black', bg='white', font=('Helvetica', 12))
    l3.place(x=15,y=180)

    #Textbox for entering password without showing it...
    e2=Entry(w, show='*')
    e2.config(highlightbackground='black',highlightthickness='.5',width=45)
    e2.place(x=20,y=210,width=270, height=30)

    #Displaying default password...
    l4=Label(w,text='Default password : password', bg='white', font=('Arial',8))
    l4.place(x=15,y=240)

    #A label showing a well wish which can be later used for showing error message as stated above...
    l5=Label(w,text='Have a nice day!', bg='white', font=('Helvetica', 12))
    l5.place(x=95, y=355)

    #Creating a button for submitting the credentials and binding it with enter button for convenience...
    b1=Button(w,text='Login', bg='#0da2ff', fg='white', relief=FLAT, width=20, height=3, command= lambda:verify(None), font=('Helvetica', 15, 'bold'))
    root.bind('<Return>', verify)
    b1.config(cursor='heart')
    b1.place(x=20, y=300, width=270, height=35)

    #Running the mainloop...
    root.mainloop()
    
#################################################################################################
###---------------------Parked Vehicles Module------------------------------------------------###
###---------------------------------------By:- Milandeep Singh--------------------------------###
#################################################################################################

#Module to display all the vehicles currently parked in the parking...
def parkedVehicles():
    #Function to show or refresh the contents...
    def show():
        
        #Code for clearing the table(the treeview widget)...
        for i in listBox.get_children():
            listBox.delete(i)

        #getting all the records from parked_vehicles table...
        myC.execute("Select * from vehicles_parked order by entry")
        #Fetching the records from the cursor and inserting it into the listBox...
        for rec in myC.fetchall():
            listBox.insert('','end', values=rec)



    #Creating root as object of Tk class of tkinter...
    root = Tk()
    #Setting title
    root.title('Parked Vehicles Info')
    #Setting geometry of root...
    root.geometry('900x750+300+10')
    #Setting background color of the root
    root.config(background='#0da2ff')
    #Configuring root to occupy all available space, so that we can arrange other things in accurate center...
    root.grid_rowconfigure(0, weight=1)
    root.grid_columnconfigure(0, weight=1)
    #Setting minimum size of the window so as to retain imp. things on the screen always...
    root.minsize(550,725)

    #Creating a canvas inside the root to put everything inside it...
    trunk=Canvas(root, background='#fff', highlightthickness=0)
    trunk.grid()

    #Creating main heading inside the trunk  with desirable visual properties...
    heading = Label(trunk, text='---- Parked Vehicles Info ----', font=('bodoni mt black', 23,'bold'), 
                    background='white', foreground='#000')
    heading.grid(row=0, column=0, columnspan=5, padx=30, pady=(10,25));

    #Exit button is placed on the top right corner for giving the ease of closing the window to the user...
    b0=Button(trunk,text='×',bg='white', relief=FLAT, width=1, height=1, command=root.destroy, font=('Helvetica', 20, 'bold'))
    b0.grid(row=0, column=5, sticky='NE', ipadx=5)
    b0.config(highlightthickness='0', cursor='hand2')

    #Creating refresh button with required attributes calling show function...
    b1=Button(trunk,text='↻',bg='white', relief=FLAT, width=1, height=1, command=show, font=('Helvetica', 20, 'bold'))
    b1.grid(row=0, column=0, sticky='NW', ipadx=5)
    b1.config(highlightthickness='0', cursor='hand2')


    # create Treeview with 7 columns...
    cols = ('Vehicle No.', 'Vehicle Type', 'Model', 'Color', 'Driver Name', 'Mob no.','Entry Timestamp')
    listBox = ttk.Treeview(trunk, columns=cols, show='headings')
    
    #setting all the 7 columns with required widths and minimum width to which they can be collapsed...
    listBox.heading('Vehicle Type')
    listBox.column('Vehicle Type', minwidth=0, width=100) 
    listBox.heading('Vehicle No.')
    listBox.column('Vehicle No.', minwidth=0, width=70)
    listBox.heading('Model')
    listBox.column('Model', minwidth=0, width=100)
    listBox.heading('Color')
    listBox.column('Color', minwidth=0, width=90)
    listBox.heading('Driver Name')
    listBox.column('Driver Name', minwidth=0, width=100)
    listBox.heading('Mob no.')
    listBox.column('Mob no.', minwidth=0, width=80)
    listBox.heading('Entry Timestamp')
    listBox.column('Entry Timestamp', minwidth=0, width=120)




    # set column headings for the table from tuple 'cols' using for loop...
    for col in cols:
        listBox.heading(col, text=col)    
    
    #griding the listbox onto the screen in second row(row 1), first column(0), with a columnspan=5...
    listBox.grid(row=1, column=0, columnspan=5, padx=(20,0),  pady=(0,20))
    
    # attach listbox to scrollbar by establishing 2 way communication between the two...
    verscrlbar = ttk.Scrollbar(trunk,  
                               orient ="vertical",
                               command = listBox.yview)
    verscrlbar.grid(row=1, column=5, sticky='NS', padx=(0,20), pady=(0,20))
    listBox.configure(yscrollcommand = verscrlbar.set)

    show()

    root.mainloop()
    
#################################################################################################
###---------------------Vehicle Entry Module--------------------------------------------------###
###---------------------------------------By:- Milandeep Singh--------------------------------###
#################################################################################################

#A module that takes all the required inputs about entering vehicle through interactive window 
#and enters the data into corresponding tables and permits the vehicle to be parked inside...
def  vehicleEntry():
    #Function which is called when user clicks Enter Button...
    def verify(event):
            #Error message string initialized to an empty string...
            m=''

            #List of inputs initialized to an empty list...
            rec=[]

            #Show corresponding messages if :-------

            #No vehicle is selected...
            if(vtype_rbutton.get()==''):
                m='Please select a vehicle type'
                messagebox.showerror('Error',m)

            #else if vehicle number is not proper 8 digit and vehicle selected is not cycle or rickshaw..
            elif (len(vno_entry.get())<7 or not vno_entry.get().isalnum() or not vno_entry.get()[:2].isalpha()):
                m='Enter vehicle number which is atleast 7 digit no. starting with two characters of state code without spaces and special characters'
                messagebox.showerror('Error',m)


            #else if vehicle model is not given...
            elif(vmodel_entry.get()==''):
                m='Please enter the vehicle model'
                messagebox.showerror('Error',m)

            #else if vehicle color is not specified
            elif(vcolor_entry.get()==''):
                m='Please specify the prominent vehicle color'
                messagebox.showerror('Error',m)

            #else if driver's name is not entered...
            elif(dname_entry.get()==''):
                m="Please enter the driver's name"
                messagebox.showerror('Error',m)

            #else if mobile number is not 10 digit
            elif(len(dmob_entry.get())!=10):
                m="Please enter 10 digit mobile number without country code(+91)"
                messagebox.showerror('Error',m)

            #else if mobile number has non-numeric characters...
            elif(not dmob_entry.get().isdigit()):
                m='Mobile number does not contain spaces, characters or special symbols...'
                messagebox.showerror('Error',m)

            #else if the id proofs are not checked...
            elif not idChecked_var.get():
                m='Please verify the id proofs...'
                messagebox.showerror('Error',m)

            #If there was no message i.e. there was no error, put the data entered into a list...
            if(m==''):
                rec=[ vno_entry.get().capitalize() if vno_entry.get() != '' else 'NULL',vtype_rbutton.get(), vmodel_entry.get(), 
                     vcolor_entry.get(), dname_entry.get().title(), dmob_entry.get(),
                     date_chosen.get(), (time_chosen['hrs'],time_chosen['min'])]

                #In case, no error is there, reconfirm the inputs from user...
                ch=messagebox.askquestion('Are you sure?',
                                      '''Please check if the following information is correct

         Vehicle number  :             %s
         Vehicle type:        %s
         Vehicle model :          %s
         Vehicle color :            %s
         Driver's name :           %s
         Mobile number :        %s
         Date of entry :           %s
         Time of entry :           %s'''
                        %(rec[0], rec[1], rec[2], rec[3], rec[4], rec[5], 
                          rec[6], str(rec[7][0])+':'+str(rec[7][1])))
                
                #If only the user clicks yes for the prompt, call the function insRec()
                if ch=='yes':
                    insRec(rec)
    

    #A function responsible for further validating and pushing the record into the required tables...
    def insRec(rec):
        
        #Value of vehicle number goes into variable 'no'...
        no = vno_entry.get()
        
        #statement for fetching the record having the same vehicle number which is entered by the user into myC cursor...
        sql="select * from vehicles_parked where lower(no)='%s';"%(no.lower)
                                                                   
        #Executing the above statement...
        myC.execute(sql);
                 
        #Shifting the records to another variable...
        result=myC.fetchall();
                                                                   
        #If some row is fetched i.e. vehicle with same number is already present...
        if result!=[]:
            #Show corresponding error message and return the control back from function...
            messagebox.showerror('Duplicacy error', 'Vehicle number is already present...')
            return

        #Statement for inserting vehicle details into the database in the otherwise case...                                                           
        sql_stuff=("INSERT INTO vehicles_parked VALUES(%s, %s, %s, %s, %s, %s, %s)")
            
        #Splitting the date inputted by user for proper formatting...                                                          
        date=rec[6].split('/')

        #Executing the command for inserting records into vehicles_parked table after formatting the things properly...
        myC.execute(sql_stuff, (rec[0], rec[1], rec[2], rec[3], rec[4], rec[5],
                        ((date[2] if len(date[2])==4 else ('20'+date[2]))+'-'+(date[0] if len(date[0])==2 else ('0'+date[0])) + '-'+ (date[1] if len(date[1])==2 else('0'+ date[1]))+' '+(str(rec[7][0]) if len(str(rec[7][0]))==2 else ('0' + str(rec[7][0]))) +':'+(str(rec[7][1]) if(len(str(rec[7][1]))==2) else ('0'+str(rec[7][1])))+':00')))
        
        #Executing the command for inserting records into vehicles_parked table after formatting the things properly...
        myC.execute("Insert into vehicle_log(no, type , model, color, driver , mobile , entry) values(%s, %s, %s, %s, %s, %s, %s)",(rec[0], rec[1], rec[2], rec[3], rec[4], rec[5],
                        ((date[2] if len(date[2])==4 else ('20'+date[2]))+'-'+(date[0] if len(date[0])==2 else ('0'+date[0])) + '-'+ (date[1] if len(date[1])==2 else('0'+ date[1]))+' '+(str(rec[7][0]) if len(str(rec[7][0]))==2 else ('0' + str(rec[7][0]))) +':'+(str(rec[7][1]) if(len(str(rec[7][1]))==2) else ('0'+str(rec[7][1])))+':00')))
        
        #Commit the operation in the database
        mydb.commit()
        
        #Show a message that vehicle has been entered successfully...
        messagebox.showinfo('Successful', 'Vehicle entered successfully')
                                                                   
        #Close the current window for clearing all te fields...
        root.destroy()
        
        #open the same window again for new entry...                                                         
        vehicleEntry()
    
    #A function which takes care of the date and time to be shown in the repective labels on the left and right side of the heading repectively...
    def date_time():
        #Get the current time...                                                           
        dt=datetime.datetime.now()
        #Update the labels of date and time...                                                           
        date_label['text']=("%s/ %s/ %02s" % (dt.month, dt.day, str(dt.year)[-2::1]))
        time_label['text']=("%02d: %02d: %02d" % (dt.hour, dt.minute, dt.second))
        #Call the function every second again...
        root.after(1000, date_time)

    #A function which takes care of enabling and disabling of the choose date and time button with respect to use system date and time checkbutton...                                                                  
    def switch():
        #If the checkbox is disabled                                                          
        if(useSysTime_var.get()==0):
            #Not disable ( Enable ) the button for custom choice...                                                       
            chooseDateTime_button.state(["!disabled"])
        #otherwise i.e. if the check box is enabled...                                                           
        else:
            #get the current date and time from the system and assign them to date_chosen and time_chosen variables respectively...
            dt=datetime.datetime.now()
            date_chosen.set("%d/%d/%d" % (dt.month, dt.day, dt.year))
            time_chosen['hrs']=dt.hour
            time_chosen['min']=dt.minute
            #Disable the button...                                                       
            chooseDateTime_button.state(["disabled"])

    #This function is invoked when user attempts to choose custom date and time for entering the vehicle... 
    def choose_date_time():
        top = Toplevel(root)

        #This function stores the date and time chosen by the user when user clicks ok...
        def store():
            #date is gotten from dateentry widget...
            date_chosen.set(date.get())
            #Custom date and time chosing window is closed...
            top.destroy()
            #time is set from spinbox widget...
            time_chosen['hrs']=time.hourstr.get()
            time_chosen['min']=time.minstr.get()

        #Label asking for choosing a date...
        ttk.Label(top, text='Choose date').pack(padx=10, pady=10)

        #creating and packing dateentry widget with suitable options...
        date = DateEntry(top)
        date = DateEntry(top, width=12, background='darkblue',
                        foreground='white', borderwidth=2, variable=date_chosen)
        date.pack(padx=10, pady=10)
        date.set_date(date_chosen.get())

        #tkinter does not have any default widget for chosing time...
        #So, this class abstracts the spinbox widget to work as time chosing widget...
        class CustomTime(tk.Frame):
            #Constructor taking it's master and time in hrs and minutes to show by default...
            def __init__(self,parent,hrs,min):
                #Initializes the master...
                super().__init__(parent)
                #Validating and setting the default no. of hours...
                self.reg=self.register(self.hour_valid)
                self.hourstr=tk.StringVar(self,hrs)
                
                #Validating and keeping the hours in the range of 0 to 23...
                self.hour = tk.Spinbox(self,from_=0,to=23,wrap=True,validate='focusout',validatecommand=(self.reg,'%P'),invalidcommand=self.hour_invalid,textvariable=self.hourstr,width=2)
                self.reg2=self.register(self.min_valid)
                
                #Similarly allowing the user to select no. of minutes...
                self.minstr=tk.StringVar(self,min)
                self.min = tk.Spinbox(self,from_=0,to=59,wrap=True,validate='focusout',validatecommand=(self.reg2,'%P'),invalidcommand=self.min_invalid,textvariable=self.minstr,width=2)
                
                #Gridding both the spinbox widgets...
                self.hour.grid()
                self.min.grid(row=0,column=1)
                
            #The function which sets the hours back to 10 if user tries to enter invalid input inside that...
            def hour_invalid(self):
                self.hourstr.set('10')
                
            #The function which checks if the hours provided are valid or not...
            def hour_valid(self,input):
                #If the input is numeric, integral, in the range of 0 to 23 and it's length in range of 1 to 2...
                if (input.isdigit() and int(input) in range(24) and len(input) in range(1,3)):
                    #The boolean variable valid is initialized to True...
                    valid = True
                    
                #Otherwise if it is not valid...
                else:
                    #Set valid to False...
                    valid = False
                #If the variable valid is False, it invokes necessary operations...
                if not valid:
                    self.hour.after_idle(lambda: self.hour.config(validate='focusout'))
                #Returns the boolean variable...
                return valid
            
            #The function which corrects the minutes if they are set invalid...
            def min_invalid(self):
                self.minstr.set('30')
                
            #This method try to keep the minutes valid which are passed as a parameter...
            def min_valid(self,input):
                #If the input is numeric, integral, in the range of 0 to 59 and it's length in range of 1 to 2...
                if (input.isdigit() and int(input) in range(60) and len(input) in range(1,3)):
                    #Set valid boolean to True
                    valid = True
                #Otherwise, 
                else:
                    #set it to False
                    valid = False
                    
                #If it is not valid, configure it to focus out which in turn will validate it...
                if not valid:
                    self.min.after_idle(lambda: self.min.config(validate='focusout'))
                #Return the flag...
                return valid

        #Creating a ttk label asking from the user about the custom time he/she wants to set...
        ttk.Label(top, text='Choose time').pack(padx=10, pady=10)
        
        #Creating the object of above-mentioned class...
        time = CustomTime(top,time_chosen['hrs'],time_chosen['min'])
        #Packing it...
        time.pack(pady=10)
        
        #Creating a button which stores the info entered in global variables and closes the top window...
        ttk.Button(top, text="ok", command=store).pack()


    #Creating a root window from the object...
    root=Tk()
    #Setting the title of the window...
    root.title('Vehicle Entry')
    #Setting the geometry and position of the window...
    root.geometry('600x750+400+10')
    #Setting background color of the window...
    root.config(background='#0da2ff')
    #Configuring the root window to expand fully in the given area so that we can keep all other widgets in the center...
    root.grid_rowconfigure(0, weight=1)
    root.grid_columnconfigure(0, weight=1)
    #Setting minimum size of the window...
    root.minsize(550,725)

    #creating a variable to store the vehicle type as selected in the radiobutton...
    vtype_rbutton = StringVar()
    vtype_rbutton.set('')

    #Variable for storing the truth value of the checkbutton wherein the user selects the fact that he wants to use system date and time or not...
    useSysTime_var = IntVar()
    useSysTime_var.set(1)

    #variable for storing the fact that the id proofs are verified or not...
    idChecked_var = IntVar()
    idChecked_var.set(0)

    #Creating a date_chosen variable as a StringVar variable...
    date_chosen = StringVar()
    
    #Loading the current date and time from system into dt variable...
    dt=datetime.datetime.now()
    
    #Setting the value of date chosen variable from the date gotten from dt variable by default...
    date_chosen.set("%s/%s/%s" % (dt.month, dt.day, dt.year))
    #Setting the time chosen variable from the current date and time by default...
    time_chosen={'hrs':dt.hour, 'min':dt.minute}

    #Creating a canvas trunk to put everything other inside it...
    trunk=Canvas(root, background='#fff', highlightthickness=0)
    #Displaying the canvas on the screen...
    trunk.grid()

    #Creating the main heading label with suitable options...
    heading = Label(trunk, text='-------Vehicle Entry-------', font=('bodoni mt black', 25,'bold'), 
                    background='white', foreground='#000')
    
    #Placing the heading in the first row and first column of the trunk canvas having columnspan=6...
    heading.grid(row=0, column=0, columnspan=6, padx=30, pady=(10,25))


    #Creating an exit button in the canvas trunk with suitable thematic options...
    b0=Button(trunk,text='×',bg='white', relief=FLAT, width=1, height=1, command=root.destroy, font=('Helvetica', 15, 'bold'))
    #gridding it in the top right corner of the window...
    b0.grid(row=0, column=4, sticky='NE', ipadx=5)
    #Setting it's thickness to zero and the cursor type to 'hand2'...
    b0.config(highlightthickness='0', cursor='hand2')

    #Label asking the user to select an option from radiobutton options with suitable visual effects...
    sel1 = Label(trunk, text='Choose vehicle type : ', font=('helvetica', 15), background='white')
    sel1.grid(row=1, column=0,  padx=5, pady=(10,0), columnspan=6, sticky='NW')

    #Creating a PhotoImage object using the 'cycle.png' image...
    cycle = PhotoImage(file = r"cycle.png", master=trunk) 
    
    #creating a subsample from the object for setting the size of the image...
    cycle = cycle.subsample(6, 6)

    #Creating an object of ttk.Style...
    s=ttk.Style()
    #Configure it to keep white as the background color for all other radiobuttons...
    s.configure('TRadiobutton', background='white')

    #Setting the first radiobutton having the image described by object : cycle...
    r1=ttk.Radiobutton(trunk, image=cycle, value='Cycle/ Rickshaw', variable=vtype_rbutton)
    #Placing it in the second row using grid with required padding...
    r1.grid(row=2, column=0, padx=(50,10), pady=4) 

    #Creating a PhotoImage object using the 'motorcycle.png' image...
    bike = PhotoImage(file = r"motorcycle.png", master=trunk) 
    #creating a subsample from the object for setting the size of the image...
    bike = bike.subsample(6, 6)

    #Setting the first radiobutton having the image described by object : bike...
    r2=ttk.Radiobutton(trunk, image=bike, value='Bike/ Scooter', variable=vtype_rbutton)
    #Placing it in the second row using grid with required padding...
    r2.grid(row=2, column=1, padx=10, pady=4, sticky = 'E') 

    #Creating a PhotoImage object using the 'auto.png' image...
    threeW = PhotoImage(file = r"auto.png", master=trunk) 
    #creating a subsample from the object for setting the size of the image...
    threeW = threeW.subsample(7, 7)

    #Setting the first radiobutton having the image described by object : threeW...
    r3=ttk.Radiobutton(trunk, image=threeW, value='Auto-Rickshaw', variable=vtype_rbutton)
    #Placing it in the second row using grid with required padding...
    r3.grid(row=2, column=2, padx=10, pady=4) 

    #Creating a PhotoImage object using the 'car.png' image...
    car = PhotoImage(file = r"car.png", master=trunk) 
    #creating a subsample from the object for setting the size of the image...
    car = car.subsample(5, 5)

    #Setting the first radiobutton having the image described by object : car...
    r4=ttk.Radiobutton(trunk, image=car, value='Car', variable=vtype_rbutton)
    #Placing it in the second row using grid with required padding...
    r4.grid(row=2, column=3, padx=10, pady=4) 

    #Creating a PhotoImage object using the 'heavy.png' image...
    heavy = PhotoImage(file = r"heavy.png", master=trunk) 
    #creating a subsample from the object for setting the size of the image...
    heavy = heavy.subsample(6, 6)

    #Setting the first radiobutton having the image described by object : heavy...
    r5=ttk.Radiobutton(trunk, image=heavy, value='Bus/ Truck', variable=vtype_rbutton)
    #Placing it in the second row using grid with required padding...
    r5.grid(row=2, column=4, padx=(10,50), pady=4) 

    #Creating a label asking for vehicle number and placing it at a proper grid condition...
    vno_label=Label(trunk,text="Vehicle number : ", fg='black', bg='white', font=('Helvetica', 15))
    vno_label.grid(row=3, column=0, columnspan=6, sticky=W, padx=10, pady=0)

    #Creating, configuring and gridding Entry widget for taking input of vehicle number...
    vno_entry=Entry(trunk)
    vno_entry.config(highlightbackground='black',highlightthickness='.5', width = 45, font = ('Helvetica', 15))
    vno_entry.grid(row=4, column=0, columnspan=6, sticky=W, padx=13, pady=(0,4))

    #Creating a label asking for vehicle model and placing it at a proper grid condition...
    vmodel_label=Label(trunk,text="Vehicle Model : ", fg='black', bg='white', font=('Helvetica', 15))
    vmodel_label.grid(row=5, column=0, columnspan=6, sticky=W, padx=10, pady=0)

    #Creating, configuring and gridding Entry widget for taking input of vehicle model...
    vmodel_entry=Entry(trunk)
    vmodel_entry.config(highlightbackground='black',highlightthickness='.5', width = 45, font = ('Helvetica', 15))
    vmodel_entry.grid(row=6, column=0, columnspan=6, sticky=W, padx=13, pady=(0,4))

    #Creating a label asking for vehicle color and placing it at a proper grid condition...
    vcolor_label=Label(trunk,text="Vehicle color : ", fg='black', bg='white', font=('Helvetica', 15))
    vcolor_label.grid(row=7, column=0, columnspan=6, sticky=W, padx=10, pady=0)

    #Creating, configuring and gridding Entry widget for taking input of vehicle color...
    vcolor_entry=Entry(trunk)
    vcolor_entry.config(highlightbackground='black',highlightthickness='.5', width = 45, font = ('Helvetica', 15))
    vcolor_entry.grid(row=8, column=0, columnspan=6, sticky=W, padx=13, pady=(0,4))

    #Creating a label asking for driver name and placing it at a proper grid condition...
    dname_label=Label(trunk,text="Driver name : ", fg='black', bg='white', font=('Helvetica', 15))
    dname_label.grid(row=9, column=0, columnspan=6, sticky=W, padx=10, pady=0)

    #Creating, configuring and gridding Entry widget for taking input of driver name...
    dname_entry=Entry(trunk)
    dname_entry.config(highlightbackground='black',highlightthickness='.5', width = 45, font = ('Helvetica', 15))
    dname_entry.grid(row=10, column=0, columnspan=6, sticky=W, padx=13, pady=(0,4))

    #Creating a label asking for mobile number and placing it at a proper grid condition...
    dmob_label=Label(trunk,text="Mobile no. : ", fg='black', bg='white', font=('Helvetica', 15))
    dmob_label.grid(row=11, column=0, columnspan=6, sticky=W, padx=10, pady=0)

    #Creating, configuring and gridding Entry widget for taking input of mobile number...
    dmob_entry=Entry(trunk)
    dmob_entry.config(highlightbackground='black',highlightthickness='.5', width = 45, font = ('Helvetica', 15))
    dmob_entry.grid(row=12, column=0, columnspan=6, sticky=W, padx=13, pady=(0,4))

    #Label for displaying date with appropriate options...
    date_label = Label(trunk, text='', background='#fff', foreground='#000', font=('Helvetica', 10, 'bold'), 
                       borderwidth=1, relief='solid')
    date_label.grid(row=0, column=0, sticky='SW', pady=(10,0), padx=10)

     #Label for displaying time with appropriate options...
    time_label = Label(trunk, text='', background='#fff', foreground='#000', font=('Helvetica', 10, 'bold'), 
                       borderwidth=1, relief='solid')
    time_label.grid(row=0, column=4, sticky='SE', padx=10)

    #creating a ttk.Style object...
    s=ttk.Style()
    s.configure('TCheckbutton', background='white', font = ('Helvetica', 15))

    #Creating checkbutton for allowing the user to use system time for the vehicle entry...
    useSysTime_checkbutton = ttk.Checkbutton(trunk, text='Use system date and time', variable=useSysTime_var, command= switch)
    useSysTime_checkbutton.grid(row=13, column=0, columnspan=3, pady=5, padx =13, sticky=W)

    #creating a button for allowing the user to choose custom timestamp for the vehicle entry...
    chooseDateTime_button = ttk.Button(trunk, text='Choose date and time', command=choose_date_time)
    chooseDateTime_button.grid(row=13, column=4, columnspan=3, pady=5, padx =(0,15), sticky=W)
    
    #By default, keep the button disabled...
    chooseDateTime_button.state(['disabled'])

    #Creating a check button for ensuring that the id proofs have been checked...
    idChecked_checkbutton = ttk.Checkbutton(trunk, text='Id proofs verified?', variable=idChecked_var)
    idChecked_checkbutton.grid(row=14, column=0, columnspan=3, pady=(5,10), padx =13, sticky=W)

    #Create an enter button for submitting all the information entered by the user...
    enter=Button(trunk,text='Enter', bg='#0da2ff', fg='white', relief=FLAT, width=41, height=1, 
                 font=('Helvetica', 15, 'bold'), cursor='hand2', command= lambda: verify(None))
    
    #Bind any enter key press with the verify function i.e. function governing info submission...
    root.bind('<Return>', verify)
    
    #gridding the enter button in the last row with columnspan=6;
    enter.grid(row=15, column=0, columnspan=6, pady=(10,30))

    #Start ticking the clock and displaying date and time in their respective labels dynamically...
    date_time()
    
    #Starting the mainloop of tkinter root window...
    root.mainloop()
    
#################################################################################################
###---------------------Vehicle Exit Module---------------------------------------------------###
###---------------------------------------By:- Milandeep Singh--------------------------------###
#################################################################################################

#Module to display all the vehicles currently parked in the parking...
def vehicleExit():
    #Function to show or refresh the contents...
    def verify(event):
        #getting the vehicle number entered by the user into no varible
        no = vno_entry.get()
        #If the length on the vehicle number is not correct,
        if(len(no)<7 or len(no)>10):
            #Show error and return either a long input may cause an exception in the database...
            messagebox.showerror('Invalid', 'Vehicle number is invalid...')
            return
        
        #statement for selecting the vehicle having number entered by the user...
        sql="select * from vehicles_parked where lower(no)='%s';"%(no.lower());
        
        #Executing the above stored statement...
        myC.execute(sql);
        
        #Putting the obtained result into a variable named result...
        result=myC.fetchall();
        
        #If no row is fetched,
        if result==[]:
            #Show corresponding error that the vehicle with this no. is not present or the number is invalid...
            messagebox.showerror('Vehicle not found', 'Vehicle number is either invalid or not present...')
        #otherwise,
        else:
            #putting the first row found back into the same variable...
            result=result[0]
            
            #tS get the time entered by the user from the interface into time_chosen variable...
            tS=(time_chosen['hrs'],time_chosen['min'])
            
            #Similarly, putting the date entered into variable called date...
            date=date_chosen.get().split('/')
            
            #Correcting the format of time stamp according to MySQL's timestamp data type from the values inserted by the user...
            tS=((date[2] if len(date[2])==4 else ('20'+date[2]))+'-'+(date[0] if len(date[0])==2 else ('0'+date[0])) + '-'+ (date[1] if len(date[1])==2 else('0'+ date[1]))+' '+(str(tS[0]) if len(str(tS[0]))==2 else ('0' + str(tS[0]))) +':'+(str(tS[1]) if(len(str(tS[1]))==2) else ('0'+str(tS[1])))+':00')
            
            #t1 stores the entry time fetched from the database...
            t1=result[6]
            
            #t2 stores the exit time chosen/ selected time by the user...
            t2=tS
            
            #Calculate the difference between the two times using a MySQL query...
            myC.execute("Select timestampdiff(minute,%s,%s) from dual;", (t1, t2))
            
            #getting the time gotten from the above stated query into tDiff variable
            tDiff=myC.fetchall()[0][0]
            
            #If the time difference is negative i.e. the exit time is less than the entry time,
            if tDiff<0:
                #Show a messagebox warning that the chosen time is not correct and ask the user to chose the time after the time at which the vehicle was entered...
                messagebox.showwarning('Warning', 'Choose time after %s'%t1)
                #and return the control back...
                return
            
            #If everything is fine then show the user all the details fetched from the database with the exit timestamp for verification...
            ch=messagebox.askquestion('Warning','''Do you want to exit the following vehicle?

         Vehicle number  :             %s
         Vehicle type:        %s
         Vehicle model :          %s
         Vehicle color :            %s
         Driver's name :           %s
         Mobile number :        %s
         Timestamp of entry :       %s
         At timestamp of exit : %s'''
                        %(result[0], result[1], result[2], result[3], result[4], result[5], 
                          result[6], tS))
                
            #If user clicks yes, then...
            if ch=="yes":
                #Fire a query for deleting the selected vehicle from the database...
                sql="delete from vehicles_parked where lower(no)='%s';"%(no.lower());
                myC.execute(sql)
                
                #Also update the vehicle log table with the exit time stamp and opposed to already prevalant NULL value...
                sql="UPDATE vehicle_log  set ext='%s' WHERE lower(no)='%s' and ext is NULL;"%(tS, no.lower())
                myC.execute(sql)
                
                #Calculate the no. of hours, the vehicle was parked with it's ceil value...
                hrs=math.ceil(tDiff/60)
                
                #Store the vehicle type in the v_type variable...
                v_type=result[1]
                
                #Calculate the bill by multiplying the price from the price_per_hour table and the no. of hours calculated previously...
                bill = price_per_hour[v_type]*hrs
                
                #Show the bill and ask if the bill is paid or not...
                ch=messagebox.askquestion('Bill', """
Time for which the %s was parked(ceil value) : %s hrs
Price per hour : Rs.%s 
Total Amount Payable : Rs. %s

Bill paid???
"""%(v_type, hrs, price_per_hour[v_type], bill))
                
                #If the bill is paid,
                if ch=='yes':
                    
                    #Show a message that the vehicle has entered successfully...
                    messagebox.showinfo('Successful', 'Vehicle exit successful')
                    
                    #Commit the database changes...
                    mydb.commit()
                    
                    #Close the current window...
                    root.destroy()
                    
                    #and return...
                    return
                #Otherwise, if the bill is not paid,
                else:
                    #Show a warning that the vehicle exit has been cancelled...
                    messagebox.showwarning('Aborted!!!', 'Vehicle exit cancelled!!!')
                    
                    #Rollback the changes in the database...
                    mydb.rollback()
                
        
                
    #A function which takes care of enabling and disabling of the choose date and time button with respect to use system date and time checkbutton...                                                                  
    def switch():
        #If the checkbox is disabled                                                          
        if(useSysTime_var.get()==0):
            #Not disable ( Enable ) the button for custom choice...                                                       
            chooseDateTime_button.state(["!disabled"])
        #otherwise i.e. if the check box is enabled...                                                           
        else:
            #get the current date and time from the system and assign them to date_chosen and time_chosen variables respectively...
            dt=datetime.datetime.now()
            date_chosen.set("%d/%d/%d" % (dt.month, dt.day, dt.year))
            time_chosen['hrs']=dt.hour
            time_chosen['min']=dt.minute
            #Disable the button...                                                       
            chooseDateTime_button.state(["disabled"])
            
    
    
    
    #A function which takes care of the date and time to be shown in the repective labels on the left and right side of the heading repectively...
    def date_time():
        #Get the current time...                                                           
        dt=datetime.datetime.now()
        #Update the labels of date and time...                                                           
        date_label['text']=("%s/ %s/ %02s" % (dt.month, dt.day, str(dt.year)[-2::1]))
        time_label['text']=("%02d: %02d: %02d" % (dt.hour, dt.minute, dt.second))
        #Call the function every second again...
        root.after(1000, date_time)

    
    #This function is invoked when user attempts to choose custom date and time for exiting the vehicle... 
    def choose_date_time():
        top = Toplevel(root)

        #This function stores the date and time chosen by the user when user clicks ok...
        def store():
            #date is gotten from dateentry widget...
            date_chosen.set(date.get())
            #Custom date and time chosing window is closed...
            top.destroy()
            #time is set from spinbox widget...
            time_chosen['hrs']=time.hourstr.get()
            time_chosen['min']=time.minstr.get()

        #Label asking for choosing a date...
        ttk.Label(top, text='Choose date').pack(padx=10, pady=10)

        #creating and packing dateentry widget with suitable options...
        date = DateEntry(top)
        date = DateEntry(top, width=12, background='darkblue',
                        foreground='white', borderwidth=2, variable=date_chosen)
        date.pack(padx=10, pady=10)
        date.set_date(date_chosen.get())

        #tkinter does not have any default widget for chosing time...
        #So, this class abstracts the spinbox widget to work as time chosing widget...
        class CustomTime(tk.Frame):
            #Constructor taking it's master and time in hrs and minutes to show by default...
            def __init__(self,parent,hrs,min):
                #Initializes the master...
                super().__init__(parent)
                #Validating and setting the default no. of hours...
                self.reg=self.register(self.hour_valid)
                self.hourstr=tk.StringVar(self,hrs)
                
                #Validating and keeping the hours in the range of 0 to 23...
                self.hour = tk.Spinbox(self,from_=0,to=23,wrap=True,validate='focusout',validatecommand=(self.reg,'%P'),invalidcommand=self.hour_invalid,textvariable=self.hourstr,width=2)
                self.reg2=self.register(self.min_valid)
                
                #Similarly allowing the user to select no. of minutes...
                self.minstr=tk.StringVar(self,min)
                self.min = tk.Spinbox(self,from_=0,to=59,wrap=True,validate='focusout',validatecommand=(self.reg2,'%P'),invalidcommand=self.min_invalid,textvariable=self.minstr,width=2)
                
                #Gridding both the spinbox widgets...
                self.hour.grid()
                self.min.grid(row=0,column=1)
                
            #The function which sets the hours back to 10 if user tries to enter invalid input inside that...
            def hour_invalid(self):
                self.hourstr.set('10')
                
            #The function which checks if the hours provided are valid or not...
            def hour_valid(self,input):
                #If the input is numeric, integral, in the range of 0 to 23 and it's length in range of 1 to 2...
                if (input.isdigit() and int(input) in range(24) and len(input) in range(1,3)):
                    #The boolean variable valid is initialized to True...
                    valid = True
                    
                #Otherwise if it is not valid...
                else:
                    #Set valid to False...
                    valid = False
                #If the variable valid is False, it invokes necessary operations...
                if not valid:
                    self.hour.after_idle(lambda: self.hour.config(validate='focusout'))
                #Returns the boolean variable...
                return valid
            
            #The function which corrects the minutes if they are set invalid...
            def min_invalid(self):
                self.minstr.set('30')
                
            #This method try to keep the minutes valid which are passed as a parameter...
            def min_valid(self,input):
                #If the input is numeric, integral, in the range of 0 to 59 and it's length in range of 1 to 2...
                if (input.isdigit() and int(input) in range(60) and len(input) in range(1,3)):
                    #Set valid boolean to True
                    valid = True
                #Otherwise, 
                else:
                    #set it to False
                    valid = False
                    
                #If it is not valid, configure it to focus out which in turn will validate it...
                if not valid:
                    self.min.after_idle(lambda: self.min.config(validate='focusout'))
                #Return the flag...
                return valid

        #Creating a ttk label asking from the user about the custom time he/she wants to set...
        ttk.Label(top, text='Choose time').pack(padx=10, pady=10)
        
        #Creating the object of above-mentioned class...
        time = CustomTime(top,time_chosen['hrs'],time_chosen['min'])
        #Packing it...
        time.pack(pady=10)
        
        #Creating a button which stores the info entered in global variables and closes the top window...
        ttk.Button(top, text="ok", command=store).pack()

        
        

    #Creating root as object of Tk class of tkinter...
    root = Tk()
    #Setting title for the exit window...
    root.title('Vehicle Exit')
    #Setting the geometry of the window...
    root.geometry('900x750+300+10')
    #Configure it's background color...
    root.config(background='#0da2ff')
    
    #Allow it to expand fully in the provided space
    root.grid_rowconfigure(0, weight=1)
    root.grid_columnconfigure(0, weight=1)
    #Setting the mnimum size to which the window can shrink...
    root.minsize(550,725)

    #Create a canvas inside it to put everything other inside that...
    trunk=Canvas(root, background='#fff', highlightthickness=0)
    trunk.grid()

    #Create the main heading using label widget with appropriate thematic visuals...
    heading = Label(trunk, text='---- Vehicle Exit ----', font=('bodoni mt black', 23,'bold'), 
                    background='white', foreground='#000')
    #Place it in the first row...
    heading.grid(row=0, column=0, columnspan=6, padx=30, pady=(10,25));

    #Creating an exit button and placing it on the top right corner on the screen for handy exit...
    b0=Button(trunk,text='×',bg='white', relief=FLAT, width=1, height=1, command=root.destroy, font=('Helvetica', 20, 'bold'))
    b0.grid(row=0, column=5, sticky='NE', ipadx=5)
    #Configuring it to have no border and 'hand2' cursor in it's domain...
    b0.config(highlightthickness='0', cursor='hand2')

    #Label asking for vehicle number to exit...
    vno_label=Label(trunk,text="Enter vehicle number to exit: ", fg='black', bg='white', font=('Helvetica', 15))
    vno_label.grid(row=3, column=0, columnspan=6, sticky=W, padx=10, pady=(30,0))

    #Entry widget for taking the input of the vehicle number...
    vno_entry=Entry(trunk)
    vno_entry.config(highlightbackground='black',highlightthickness='.5', width = 45, font = ('Helvetica', 15))
    #Putting it in 4th row with columnspan=6
    vno_entry.grid(row=4, column=0, columnspan=6, sticky=W, padx=13, pady=30)

    #A button for submitting the number and exit time of the vehicle to be exited and invokes verify function...
    exit=Button(trunk,text='Exit', bg='#0da2ff', fg='white', relief=FLAT, width=41, height=1, 
                 font=('Helvetica', 15, 'bold'), cursor='hand2', command= lambda: verify(None))
    
    #Binding the enter key press with the verify function...
    root.bind('<Return>', verify)
    
    #Put it in the last row of the window, again with columnspan=6...
    exit.grid(row=6, column=0, columnspan=6, pady=30)
    
    
    #Variable for storing the truth value of the checkbutton wherein the user selects the fact that he wants to use system date and time or not...
    useSysTime_var = IntVar()
    #Set the checkbox to True by default...
    useSysTime_var.set(1)

    #Creating a date_chosen variable as a StringVar variable...
    date_chosen = StringVar()
    
    #Loading the current date and time from system into dt variable...
    dt=datetime.datetime.now()
    
    #Setting the value of date chosen variable from the date gotten from dt variable by default...
    date_chosen.set("%s/%s/%s" % (dt.month, dt.day, dt.year))
    #Setting the time chosen variable from the current date and time by default...
    time_chosen={'hrs':dt.hour, 'min':dt.minute}
    
    #Checkbutton for asking user if he/she wants to use system date and time...
    useSysTime_checkbutton = ttk.Checkbutton(trunk, text='Use system date and time', variable=useSysTime_var, command= switch)
    useSysTime_checkbutton.grid(row=5, column=0, columnspan=3, pady=5, padx =13, sticky=W)

    #A button which invokes a function for custom date and time selection...
    chooseDateTime_button = ttk.Button(trunk, text='Choose date and time', command=choose_date_time)
    chooseDateTime_button.grid(row=5, column=5, columnspan=3, pady=5, padx =(0,15), sticky=W)
    #Put it to disable mode by default...
    chooseDateTime_button.state(['disabled'])
    
    #A label on upper left corner of the trunk window showing realtime date as fetched from the system...
    date_label = Label(trunk, text='', background='#fff', foreground='#000', font=('Helvetica', 10, 'bold'), 
                   borderwidth=1, relief='solid')
    date_label.grid(row=0, column=0, sticky='SW', pady=(10,0), padx=10)

    #A label on the upper right corner of the trunk window displaying realtime system time dynamically...
    time_label = Label(trunk, text='', background='#fff', foreground='#000', font=('Helvetica', 10, 'bold'), 
                   borderwidth=1, relief='solid')
    time_label.grid(row=0, column=5, sticky='SE', padx=10)
    
    #Start the date_time function which starts showing realtime time...
    date_time()
    
    #Starting the mainloop...
    root.mainloop()
    
#Calling the signin function from which the further modules can be invoked and traversed...
signIn()
```
